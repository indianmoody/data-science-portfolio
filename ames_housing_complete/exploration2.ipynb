```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data/train.csv')
```

Convert 'MSSubClass' and 'MoSold' as categorical.

```python
mssubclass_mapper = {
    20 : 'mssub1',
    30 : 'mssub2',
    40 : 'mssub3',
    45 : 'mssub4',
    50 : 'mssub5',
    60 : 'mssub6',
    70 : 'mssub7',
    75 : 'mssub8',
    80 : 'mssub9',
    85 : 'mssub10',
    90 : 'mssub11',
    120 : 'mssub12',
    150 : 'mssub13',
    160 : 'mssub14',
    180 : 'mssub15',
    190 : 'mssub16'
}

mosold_mapper = {
    1 : 'january',
    2 : 'february',
    3 : 'march',
    4 : 'april',
    5 : 'may',
    6 : 'june',
    7 : 'july',
    8 : 'august',
    9 : 'september',
    10 : 'october',
    11 : 'november',
    12 : 'december'
}

df['MSSubClass'].replace(mssubclass_mapper, inplace=True)
df['MoSold'].replace(mosold_mapper, inplace=True)
```

```python
df.drop(columns=['Id','PoolQC','LowQualFinSF','WoodDeckSF',  'OpenPorchSF',  'EnclosedPorch',  '3SsnPorch',  'ScreenPorch',  'PoolArea',  'MiscVal'], inplace=True)
```

Group 'LotShape' into two categories: Reg and Irreg. Fill missing with mode.

```python
def lotshape_trans(x):
    if (x =='Reg'):
        return x
    else:
        return 'Irreg'
```
