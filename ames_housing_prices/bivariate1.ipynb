```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style("whitegrid")
```

```python
ames = pd.read_csv("data/train.csv")
ames.head()
```

### Continuous Variables

```python
list1 = ['LotFrontage','LotArea','YearBuilt','YearRemodAdd','MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF']
```

```python
# Correlation of these variables with target 'SalePrice'
ames[list1+['SalePrice']].corr()['SalePrice']
```

```python
fig2 = plt.figure(figsize=(14,14))
for i in range(1,10):
    ax = fig2.add_subplot(3,3,i)
    ax.scatter(ames[list1[i-1]],ames['SalePrice'], s=5)
    ax.set_xlabel(ames[list1[i-1]].name)
```

```python
ames['SalePrice'].name
```

```python
plt.scatter(np.log(ames['LotArea']), ames['SalePrice'], s=2)
```

```python
plt.scatter(np.log(ames['LotFrontage']), ames['SalePrice'], s=2)
```

```python
plt.scatter(np.sqrt(ames['BsmtUnfSF']), ames['SalePrice'], s=2)
```

'BsmtFinSF2' doesn't seem to affect 'SalePrice'. Even after binning.
