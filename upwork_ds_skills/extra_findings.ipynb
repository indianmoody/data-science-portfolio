```python
import plotly.graph_objs as go
```

```python
# Read csv file into pandas DataFrame
skills_df = pd.read_csv('job_postings_data.csv', index_col=0)
```

```python
skills_df.head(5)
```

## Top 10 Frequent Skills

```python
top = skills_df['Skill'].value_counts()[:10]
top_skills = top.index
top_count = top.values
top_df = skills_df[skills_df['Skill'].isin(list(top_skills))]
```

```python
top_df.loc[:,'Count Order'] = top_df['Skill'].apply(lambda x: top[x])
```

```python
top_df.head()
```

```python
top_df = top_df.sort_values(by=['Count Order','Skill'], ascending=[0,1])
top_df.head()
```

```python
extra_layout = go.Layout(
    barmode='stack',
    bargap=0.5,
    #margin = go.Margin(b=100),
    xaxis=dict(
        #showticklabels=False,
        showgrid=False,
        #zeroline=False,
        #tickangle=-45
    ),
    yaxis=dict(
        showgrid=True,
        #showline=False,
        #zeroline=False,
        #ticks='outside'
    )
)
```

```python
top_entry = top_df[top_df['Job Tier']=='Entry']['Skill']
top_inter = top_df[top_df['Job Tier']=='Intermediate']['Skill']
top_expert = top_df[top_df['Job Tier']=='Expert']['Skill']

trace0 = go.Histogram(x=top_entry, name='Entry Level', marker=dict(color='#C39BD3'))
trace1 = go.Histogram(x=top_inter, name='Intermediate', marker=dict(color='#A9CCE3'))
trace2 = go.Histogram(x=top_expert, name='Expert Level', marker=dict(color='#A3E4D7'))

data = [trace0,trace1,trace2]
```

```python
iplot(go.Figure(data=data, layout = extra_layout), show_link=False)
```
