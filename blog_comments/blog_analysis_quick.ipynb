```python
param_grid = {'C': [0.1,1,10], 'gamma': [0.1,0.01,0.001], 'kernel': ['rbf']}
```

```python
svr_scaler_y = StandardScaler()
```

```python
ytrain_svr = svr_scaler_y.fit_transform(df.iloc[:30000,-1].values.reshape(-1,1))
```
