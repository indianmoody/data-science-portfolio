```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)
```

```python
training_scores = []
test_scores = []
```

```python
for neighbors in range(1,21,2):
    knn = KNeighborsClassifier(n_neighbors=neighbors).fit(X_train, y_train)
    training_scores.append(knn.score(X_train, y_train))
    test_scores.append(knn.score(X_test, y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,21,2), training_scores, label='training', marker='.')
plt.plot(range(1,21,2), test_scores, label='test', marker='.')
plt.legend()
plt.title('Accuracy Score with with different k values of KNN')
plt.xlabel('k')
plt.ylabel('Accuracy Score')
```
