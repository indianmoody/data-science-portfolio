```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from wordcloud import WordCloud
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("data/spam.csv", encoding='ISO-8859-1')
```

```python
df.sample(5)
```

```python
df = df.iloc[:,[0,1]]
```

```python
df.columns = ['labels', 'data']
```

```python
df.head()
```

```python
df['labels'].value_counts()
```

```python
count_vectorizer = CountVectorizer(decode_error='ignore')
X_count = count_vectorizer.fit_transform(df['data'])
```

```python
tfidf_vectorizer = TfidfVectorizer(decode_error='ignore')
X_tfidf = tfidf_vectorizer.fit_transform(df['data'])
```

```python
Xtrain1, Xtest1, ytrain1, ytest1 = train_test_split(X_count, df['labels'], test_size = 0.3)
Xtrain2, Xtest2, ytrain2, ytest2 = train_test_split(X_tfidf, df['labels'], test_size = 0.3)
```

```python
model1 = MultinomialNB()
model1.fit(Xtrain1, ytrain1)
print("train score:", model1.score(Xtrain1, ytrain1))
print("test score:", model1.score(Xtest1, ytest1))
```

```python
model2 = MultinomialNB()
model2.fit(Xtrain2, ytrain2)
print("train score:", model2.score(Xtrain2, ytrain2))
print("test score:", model2.score(Xtest2, ytest2))
```

```python
def visualize(label):
    words = ''
    for msg in df[df['labels'] == label]['data']:
        msg = msg.lower()
        words += msg + ' '
    wordcloud = WordCloud(width = 1200, height = 800).generate(words)
    plt.figure(figsize=(12,8))
    plt.imshow(wordcloud)
    plt.axis('off')
```

```python
visualize('ham')
```

```python
visualize('spam')
```
